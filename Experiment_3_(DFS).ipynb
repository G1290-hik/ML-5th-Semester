```python
#@title Necessary Imports
import time
import networkx as nx
import matplotlib.pyplot as plt
```

```python
#@title Graph Class
class Graph:

    def __init__(self, vertices):
        self.V = vertices
        self.graph = [[] for _ in range(vertices)]

    def add_edge(self, u, v):
        self.graph[u].append(v)

    def dfs_util(self, v, visited):
        visited[v] = True
        print(v, end=' ')
        for neighbor in self.graph[v]:
            if not visited[neighbor]:
                self.dfs_util(neighbor, visited)

    def dfs(self, start_vertex):
        visited = [False] * self.V
        self.dfs_util(start_vertex, visited)

    def visualize(self):
        G = nx.DiGraph()
        for u in range(self.V):
            for v in self.graph[u]:
                G.add_edge(u, v)

        pos = nx.spring_layout(G)
        nx.draw(G, pos, with_labels=True, node_size=700, node_color='green', font_size=14, font_color='white', font_weight='bold', arrows=True)
        plt.show()

def measure_dfs_performance(graph, start_vertex):
    start_time = time.time()
    graph.dfs(start_vertex)
    end_time = time.time()
    print(f"\nDFS execution time: {end_time - start_time:.10f} seconds")
```

```python
#@title Implementation
g = Graph(8)
g.add_edge(0, 4)
g.add_edge(0, 5)
g.add_edge(4, 3)
g.add_edge(4, 1)
g.add_edge(5, 2)
g.add_edge(5, 6)
g.add_edge(5, 7)
start_vertex = 0
print(f"DFS starting from vertex {start_vertex}:")
measure_dfs_performance(g, start_vertex)
print("\n")
g.visualize()
```
